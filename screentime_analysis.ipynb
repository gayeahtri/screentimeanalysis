```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("Screentime-App-Details.csv")
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
fig = px.bar(data_frame = df,x="Date",y="Usage",color="App",title="usage")
fig.show()
```

```python
fig1 = px.bar(data_frame=df,x="Date",y="Notifications",color="App",title="Notifications")
fig1.show()
```

no of times an app is opened

```python
fig2 = px.bar(data_frame = df,x="Date",y="Times opened",color = "App",title="Times opened")
fig2.show()
```

relation between notifications and times opened

```python
fig3 = px.scatter(data_frame = df,x="Notifications",y="Usage",trendline="ols",title ="relation between usage and no of notifications" )
fig3.show()
```
